```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#unique screennames:
clean_df =purchase_data.dropna(how="any")
clean_df = clean_df.drop_duplicates()
stats = clean_df.loc[:, ["Gender", "SN", "Age"]]
stats = stats.drop_duplicates()
clean_df.count()
```

```python
player_count = len(cleaner_df["SN"].unique())
print(player_count)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#fomulas:
item_count = len(cleaner_df["Item ID"].unique())
avg_price = cleaner_df["Price"].mean()
purchase_count = cleaner_df["Purchase ID"].count()
total_rev = cleaner_df["Price"].sum()

#display results in DF
summary = pd.DataFrame({"Number of Unique Items" : [item_count],
                       "Average Price" : [avg_price],
                       "Number of Purchases" : [purchase_count],
                       "Total Revenue" : [total_rev]})

summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#fomula for unique demographics:
unique_players = stats.count()[0]
all_gender_count = stats["Gender"].value_counts()
all_gender_pct = (all_gender_count / unique_players) * 100

#display in gender DF
gender_summary = pd.DataFrame({"Total Count" : all_gender_count,
                       "Percentage of Players" : all_gender_pct})

gender_summary 
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
count_by_gender = clean_df.groupby(["Gender"]).count()["Price"].rename("Purchase Count")
avg_by_gender = clean_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
purchases_by_gender_total = clean_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")

gender_purchasing_behavior = pd.DataFrame({"Purchase Count" : count_by_gender,
                                          "Average Purchase Price" : avg_by_gender,
                                          "Total Purchase Value" : purchases_by_gender_total})

gender_purchasing_behavior
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bin_ranges = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 99999999999999999999]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">39" ]
stats["Age Ranges"] = pd.cut(stats["Age"], bin_ranges, labels = bin_names)

age_totals = stats["Age Ranges"].value_counts()
age_pcts = age_totals / unique_players
age_summary = pd.DataFrame({"Total Count" : age_totals,
                           "Percentage of Players" : age_pcts})

age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
clean_df["Age Ranges"] = pd.cut(stats["Age"], bin_ranges, labels = bin_names)

age_purchase_counts = clean_df.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")
age_purchase_avg = clean_df.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
age_purchase_ttl = clean_df.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")

age_purchase_summary = pd.DataFrame({"Purchase Count": age_purchase_counts,
                                     "Average Purchase Price": age_purchase_avg,
                                     "Total Purchase Value": age_purchase_ttl})

age_purchase_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_total = clean_df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")
sn_avg = clean_df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
sn_count = clean_df.groupby(["SN"]).count()["Price"].rename("Purchase Count")

sn_data = pd.DataFrame({"Purchase_Count" : sn_count,
                       "Average Purchase Price" : sn_avg,
                        "Total Purchase Value" : sn_total})

sn_data.sort_values(by=['Total Purchase Value'], ascending=False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = clean_df.loc[:,["Item ID", "Item Name", "Price"]]
item_total = items.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")
item_avg_sold = items.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_count_sold = items.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")

item_summary = pd.DataFrame({"Purchase Count": item_count_sold,
                             "Item Price": item_avg_sold,
                             "Total Purchase Value": item_total})

item_summary.sort_values(by=['Purchase Count'], ascending=False)
```
